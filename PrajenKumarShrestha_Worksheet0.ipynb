Task - 1: Create a Python program that converts between different units of measurement.

```python
def convert_units(conversion_type, value):
    """
    Convert units between length, weight, and volume.

    Parameters:
    conversion_type (str): Type of conversion
    value (float): Value to convert

    Returns:
    float: Converted value
    """

    if conversion_type == "m_to_ft":
        return value * 3.28084
    elif conversion_type == "ft_to_m":
        return value / 3.28084
    elif conversion_type == "kg_to_lbs":
        return value * 2.20462
    elif conversion_type == "lbs_to_kg":
        return value / 2.20462
    elif conversion_type == "l_to_gal":
        return value * 0.264172
    elif conversion_type == "gal_to_l":
        return value / 0.264172
    else:
        raise ValueError("Unsupported conversion type")


def unit_converter():
    try:
        print("Choose conversion type:")
        print("1: meters → feet")
        print("2: feet → meters")
        print("3: kg → pounds")
        print("4: pounds → kg")
        print("5: liters → gallons")
        print("6: gallons → liters")

        choice = input("Enter choice: ")
        value = float(input("Enter value to convert: "))

        mapping = {
            "1": "m_to_ft",
            "2": "ft_to_m",
            "3": "kg_to_lbs",
            "4": "lbs_to_kg",
            "5": "l_to_gal",
            "6": "gal_to_l"
        }

        result = convert_units(mapping[choice], value)
        print("Converted value:", result)

    except Exception as e:
        print("Error:", e)


unit_converter()
```

```python
def calculate_sum(numbers):
    """Return sum of numbers."""
    return sum(numbers)

def calculate_average(numbers):
    """Return average of numbers."""
    return sum(numbers) / len(numbers)

def find_max(numbers):
    """Return maximum number."""
    return max(numbers)

def find_min(numbers):
    """Return minimum number."""
    return min(numbers)


def list_operations():
    try:
        numbers = list(map(float, input("Enter numbers separated by space: ").split()))

        if len(numbers) == 0:
            raise ValueError("List cannot be empty")

        print("Choose operation:")
        print("1. Sum")
        print("2. Average")
        print("3. Maximum")
        print("4. Minimum")

        choice = input("Enter choice: ")

        if choice == "1":
            print("Sum:", calculate_sum(numbers))
        elif choice == "2":
            print("Average:", calculate_average(numbers))
        elif choice == "3":
            print("Maximum:", find_max(numbers))
        elif choice == "4":
            print("Minimum:", find_min(numbers))
        else:
            print("Invalid choice")

    except Exception as e:
        print("Error:", e)


list_operations()
```

**4.3 Exercise on Nested List:**
1. Flatten a Nested List:
Write a Python function that takes a nested list and flattens it into a single list, where all the elements are in a single dimension.

```python
lst = [[1, 2], [3, 4], [5]]

def flatten(lst):
    result = []
    for sub in lst:
        if isinstance(sub, list):
            result.extend(sub)
        else:
            result.append(sub)
    return result

flatten(lst)
```

2. Accessing Nested List Elements:
Write a Python function that extracts a specific element from a nested list given its indices.

```python
lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1,2]

def access_nested_element(lst, indices):
    element = lst
    for i in indices:
        element = element[i]
    return element

access_nested_element(lst, indices)
```

3. Sum of All Elements in a Nested List:
Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).

```python
lst = [[1, 2], [3, [4, 5]], 6]

def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

sum_nested(lst)
```

4. Remove Specific Element from a Nested List:
Write a Python function that removes all occurrences of a specific element from a nested list.

```python
lst = [[1, 2], [3, 2], [4, 5]]
elem = 2

def remove_element(lst, elem):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        else:
            if item != elem:
                result.append(item)
    return result

remove_element(lst, elem)
```

5. Find the Maximum Element in a Nested List:
Write a Python function that finds the maximum element in a nested list (regardless of depth).

```python
lst = [[1, 2], [3, [4, 5]], 6]

def find_max(lst):
    maximum = float('-inf')
    for item in lst:
        if isinstance(item, list):
            maximum = max(maximum, find_max(item))
        else:
            maximum = max(maximum, item)
    return maximum

find_max(lst)
```

6. Count Occurrences of an Element in a Nested List:
Write a Python function that counts how many times a specific element appears in a nested list.

```python
lst = [[1, 2], [2, 3], [2, 4]]
elem = 2

def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count

count_occurrences(lst, elem)
```

7. Flatten a List of Lists of Lists:
Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

```python
lst = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

def deep_flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

deep_flatten(lst)
```

8. Nested List Average:
Write a Python function that calculates the average of all elements in a nested list.

```python
lst =  [[1, 2], [3, 4], [5, 6]]

def average_nested(lst):
    flat = deep_flatten(lst)
    return sum(flat)/len(flat)

average_nested(lst)
```

Problem - 4: Matrix Operations:

For the following arrays:

A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);

Prove following with Numpy:

1. Prove A.A−1 = I.
2. Prove AB ̸= BA.
3. Prove (AB)T = BTAT

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1  A * A^-1
A_inv = np.linalg.inv(A)
print(np.dot(A, A_inv))
```

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 2  AB != BA
print("AB =\n", np.dot(A,B))
print("BA =\n", np.dot(B,A))
```

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 3 (AB)^T = B^T A^T
left = np.transpose(np.dot(A,B))
right = np.dot(np.transpose(B), np.transpose(A))

print(left)
print(right)
```

**10.2 Experiment: How Fast is Numpy?**

```python
import numpy as np
import time

size = 1000000

a = list(range(size))
b = list(range(size))

start = time.time()
result = [a[i] + b[i] for i in range(size)]
print("Python addition time:", time.time()-start)

a_np = np.array(a)
b_np = np.array(b)

start = time.time()
result = a_np + b_np
print("Numpy addition time:", time.time()-start)
```
